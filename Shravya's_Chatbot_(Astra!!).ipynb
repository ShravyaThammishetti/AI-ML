```python
import requests
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import os

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')

# List of Wikipedia URLs related to Space
urls = [
    "https://en.wikipedia.org/wiki/Space_exploration",
    "https://en.wikipedia.org/wiki/Aryabhata_(satellite)",
    "https://en.wikipedia.org/wiki/Chandrayaan-1",
    "https://en.wikipedia.org/wiki/Mars_Orbiter_Mission",
    "https://en.wikipedia.org/wiki/Polar_Satellite_Launch_Vehicle",
    "https://en.wikipedia.org/wiki/Geosynchronous_Satellite_Launch_Vehicle",
    "https://en.wikipedia.org/wiki/Gaganyaan",
    "https://en.wikipedia.org/wiki/GSAT",
    "https://en.wikipedia.org/wiki/Indian_National_Satellite_System",
    "https://en.wikipedia.org/wiki/RISAT",
    "https://en.wikipedia.org/wiki/Indian_Regional_Navigation_Satellite_System",
    "https://en.wikipedia.org/wiki/Cartosat",
    "https://en.wikipedia.org/wiki/Astrosat",
    "https://en.wikipedia.org/wiki/Vikram_Sarabhai"
]

# Function to scrape a Wikipedia page and return its content with headings and subheadings
def scrape_wikipedia_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')


    title = soup.find('h1', {'id': 'firstHeading'}).text
    title = re.sub(r'[\\/*?:"<>|]', "", title)  # Remove invalid characters for file names


    content = soup.find('div', {'class': 'mw-parser-output'})

    page_text = []

    # Extracting headings, subheadings, and paragraphs
    for element in content.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p']):
        if element.name.startswith('h'):
            page_text.append('\n' + element.text.strip() + '\n')
        elif element.name == 'p':
            page_text.append(element.text.strip())

    # Combine all elements into a single string
    page_text = '\n'.join(page_text)

    return title, page_text

# Preprocessing function
def preprocess_text(text):
    text = text.lower()

    # Remove punctuation and numbers
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenization
    tokens = word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return ' '.join(tokens)

# Scrape each page and save the content in a text file
for url in urls:
    title, text = scrape_wikipedia_page(url)
    filename = f"{title}.txt"

    # Save the raw text to a file
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(text)
    print(f"Saved {title} to {filename}")

    # Preprocess the text
    preprocessed_text = preprocess_text(text)
    preprocessed_filename = f"preprocessed_{title}.txt"

    # Save the preprocessed text to a new file
    with open(preprocessed_filename, 'w', encoding='utf-8') as file:
        file.write(preprocessed_text)
    print(f"Saved preprocessed content to {preprocessed_filename}")

print("Scraping and preprocessing completed.")
```
