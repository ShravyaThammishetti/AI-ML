```python
import pandas as pd
import nltk #Natural Language Toolkit --> for working with human language data.
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon
nltk.download('vader_lexicon') #VADER --> specifically designed for sentiment analysis


def map_to_range(score, lower_bound, upper_bound):
    # Map the score to the specified range
    return (score + 1) * (upper_bound - lower_bound) / 2 + lower_bound

def sentiment_score(review):
    # Initialize VADER Sentiment Intensity Analyzer
    sid = SentimentIntensityAnalyzer()

    # Calculate sentiment scores for the review
    scores = sid.polarity_scores(review) #returns a dictionary of pos ,neg ,neu & compound.

    # Normalize the compound score to the range [1, 5]
    normalized_score = map_to_range(scores['compound'],1,5)

    return round(normalized_score)


# Load CSV file into a DataFrame
file_path = '/content/output_filt2.csv'
df = pd.read_csv(file_path)

df['Sentiment Score'] = df['Review'].apply(sentiment_score)

# Save the updated DataFrame back to the CSV file
df.to_csv('outputsent2.csv', index=False)
```

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('punkt')  #Punkt tokenizer, a pre-trained unsupervised machine learning model for tokenizing text into sentences.
nltk.download('punkt_tab') #Punkt_tab , specifically to handle punctuation
nltk.download('vader_lexicon') #It maps words and phrases to sentiment intensity scores based on human-rated emotional valence.

def map_to_range(score, lower_bound, upper_bound):
    # Map the score to the specified range
    return (score + 1) * (upper_bound - lower_bound) / 2 + lower_bound

def analyze_sentence_sentiment(sentence):
    # Initialize VADER Sentiment Intensity Analyzer
    sid = SentimentIntensityAnalyzer()

    # Check if the keyword is mentioned in the review
    if 'comfort' in sentence.lower():
        # Calculate sentiment scores for the review
        scores = sid.polarity_scores(sentence)
        return scores['compound']

def analyze_review_sentiments(review):
    # Tokenize the review into sentences using NLTK
    sentences = nltk.sent_tokenize(review)
    sentence_sentiments=[]
    # Analyze sentiment for each sentence
    for sentence in sentences:
      k=analyze_sentence_sentiment(sentence)
      if k:
        sentence_sentiments.append(k)

    # Calculate the overall sentiment for the entire review based on the average sentiment of sentences
    if len(sentence_sentiments) > 0:
        overall_sentiment = max(sentence_sentiments, key=abs)
        mapped_score = map_to_range(overall_sentiment, 1, 5)

        # Round the mapped score to the nearest integer
        overall_sentiment = round(mapped_score)


    else:
        overall_sentiment = 'NaN'  # Default to null if there are no sentences

    return overall_sentiment

# Load CSV file into a DataFrame
file_path = '/content/outputsent2.csv'
df = pd.read_csv(file_path)

# Add a new column 'Comfort score' based on sentiment analysis for each sentence
df['Comfort score'] = df['Review'].apply(analyze_review_sentiments)

df.to_csv('da21.csv', index=False)
```

```python
# Initialize Rake
rake_nltk_var = Rake() #used for automatic keyword extraction
```

```python
df = pd.read_csv('/content/da25.csv')
```

```python
# Add  new column for positive word of that row
df['Positive_Phrase'] = ''

# Add a new column for  negative word of that row
df['Negative_Phrase'] = ''
```

```python
def extract_phrases(text):
    # Perform sentiment analysis on the text
    text_sentiment = TextBlob(text).sentiment.polarity

    # Extract keywords using Rake (Rapid Automatic Keyword Extraction)
    rake_nltk_var.extract_keywords_from_text(text)
    keyword_extracted = rake_nltk_var.get_ranked_phrases()

    # Calculate sentiment scores for each keyword
    keyword_sentiments = [TextBlob(phrase).sentiment.polarity for phrase in keyword_extracted]

    # Combine keywords with their sentiment scores
    keyword_sentiment_pairs = list(zip(keyword_extracted, keyword_sentiments))

    # Sort keywords based on sentiment scores
    sorted_keywords = sorted(keyword_sentiment_pairs, key=lambda x: x[1], reverse=True)

    # Find the most positive and most negative phrases
    positive_phrases = [phrase for phrase, sentiment in sorted_keywords if sentiment > 0.7][:1]
    negative_phrases = [phrase for phrase, sentiment in sorted_keywords if sentiment < -0.2][:1]

    # Return one positive and one negative phrase
    return positive_phrases, negative_phrases

# Example usage:
for index, row in df.iterrows():
    text = row['Review']  # column name containing text
    positive_phrase, negative_phrase = extract_phrases(text)
    print(f"Positive Phrase: {positive_phrase}, Negative Phrase: {negative_phrase}")
    # Save positive and negative phrases to new columns
    df.at[index, 'Positive_Phrase'] = ', '.join(positive_phrase)
    df.at[index, 'Negative_Phrase'] = ', '.join(negative_phrase)
```
